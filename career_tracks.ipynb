```python
import pandas as pd

info = pd.read_csv("tables/career_track_info.csv")
info
```

```python
enroll = pd.read_csv("tables/career_track_enrollments.csv")
enroll
```

```python
df = pd.DataFrame(data=[], columns=["action", "track_id", "n_students"])

df["track_id"] = [1, 2, 3] * 5
df
```

```python
df["action"] = ["track enrollments"] * 3 + ["course exam attempts"] * 3 + ["course exam passed"] * 3 + ["final exam attempts"] * 3 + ["final exam passed"] * 3
df
```

```python
for i in range(3):
    df.loc[i, "n_students"] = len(enroll[enroll.track_id == i+1])

df
```

```python
exams = pd.read_csv("tables/student_exams.csv")
exams
```

```python
exam_info = pd.read_csv("tables/exam_info.csv")
exam_info
```

```python
course_exam_info = exam_info[exam_info.exam_category == 2].drop(["track_id", "exam_duration"], axis=1)
course_exam_info
```

```python
exams_ext = pd.merge(exams, course_exam_info, on="exam_id", how="left")
exams_ext
```

```python
course_exams = exams_ext[exams_ext.exam_category == 2]
course_exams
```

```python
course_exams = pd.merge(course_exams, enroll, on="student_id", how="left")
course_exams
```

```python
course_exams.date_exam_completed = pd.to_datetime(course_exams.date_exam_completed)
course_exams.date_enrolled = pd.to_datetime(course_exams.date_enrolled)
course_exams.dtypes
```

```python
def track_course_exam(track_id, only_pass=False):
    df = course_exams.dropna()
    track_courses = info[info.track_id == track_id].course_id.astype(str)
    df = df[df.track_id == track_id]
    if only_pass:
        df = df[df.exam_passed == 1]
    return df[(df.course_id.isin(track_courses)) & (df.date_enrolled <= df.date_exam_completed)]

for n, i in enumerate(range(3, 6)):
    df.loc[i, "n_students"] = track_course_exam(n+1).student_id.nunique()

for n, i in enumerate(range(6, 9)):
    df.loc[i, "n_students"] = track_course_exam(n+1, True).student_id.nunique()

df
```

```python
track_exam_info = exam_info[exam_info.exam_category == 3]

track_exams = pd.merge(exams, track_exam_info, on="exam_id", how="left").dropna()
track_exams.track_id = track_exams.track_id.astype(int)
track_exams.dtypes
```

```python
for n, i in enumerate(range(9, 12)):
    df.loc[i, "n_students"] = track_exams[track_exams.track_id == n+1].student_id.nunique()

df
```

```python

for n, i in enumerate(range(12, 15)):
    df.loc[i, "n_students"] = track_exams[(track_exams.exam_passed == 1) & (track_exams.track_id == n+1)].student_id.nunique()

df
```
