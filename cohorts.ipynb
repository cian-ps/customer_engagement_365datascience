```python
import pandas as pd

df = pd.read_csv("data/engagement.csv")
df
```

```python
cohorts = df.groupby(["student_id", "paid"], as_index=False).engagement_date.min()
cohorts
```

```python
cohorts = cohorts.rename(columns={"engagement_date": "cohort"})
cohorts
```

```python
df = pd.merge(df, cohorts, on=["student_id", "paid"])
df
```

```python
df.cohort = pd.to_datetime(df.cohort).dt.strftime("%Y-%m")
df
```

```python
df["period"] = pd.to_datetime(df.engagement_date).dt.month - pd.to_datetime(df.cohort).dt.month
df
```

```python
df.to_csv("data/cohorts.csv", index=False)
```
